```python
#indexing for 1-d matrix
print(a[2])
print(a[:])
print(a[0::2])
print(a[1::2])
```

```python
#indexing for 2-d matrix
print(b[0,1])
print(b[0,:])#prints the 1st row
print(b[:,1])#prints the 2nd column
print(b[0:2:1,0])#[start_index:end_index:step_count]
```
